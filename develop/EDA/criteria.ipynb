```python
import numpy as np
from sklearn import preprocessing
```

```python
# サンプル配列を作成
data1 = np.array([10, 25, 50, 75, 100]).reshape(-1, 1)
```

```python
data2 = np.array([1,2,3,3,6]).reshape(-1, 1)
```

```python
data3 = np.array([1022,1233,6022,2222,1232]).reshape(-1, 1)
```

```python
# サンプル配列を作成
data = data2

# 最大値・最小値
print(f"元データ: {data.flatten()}")
print(f"最大値: {data.max()}")
print(f"最小値: {data.min()}")

# 元データの最小単位（整数データなら1）
min_unit = 1

# 0-1に圧縮（MinMaxScaler）
scaler = preprocessing.MinMaxScaler()
data_normalized = scaler.fit_transform(data)

print(f"正規化後: {data_normalized.flatten()}")

# 正規化後の最小メモリ（最小単位を正規化した値）
data_range = data.max() - data.min()
normalized_min_unit = min_unit / data_range

print(f"元データの範囲: {data_range}")
print(f"元データの最小単位: {min_unit}")
print(f"正規化後の最小メモリ: {normalized_min_unit}")
```
