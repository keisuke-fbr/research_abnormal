```python
import pandas as pd
import os
```

```python
# 結合したいファイルが格納されているディレクトリ
input_dir = "result"  # ファイルのディレクトリを指定
```

```python
# 出力ファイル名
output_file = "combined_anomaly_scores.csv"
```

```python
# 結合するファイルのリストを取得
file_list = [os.path.join(input_dir, f) for f in os.listdir(input_dir) if f.endswith('.csv')]

# ファイルの読み込みと結合
combined_df = pd.DataFrame()  # 空のデータフレームを作成
```

```python
for file in file_list:
    print(f"ファイルを読み込み中: {file}")
    temp_df = pd.read_csv(file)  # 各CSVを読み込む
    combined_df = pd.concat([combined_df, temp_df], ignore_index=True)  # データを結合
```

```python
# 結合結果を新しいファイルとして保存
combined_df.to_csv(output_file, index=False)

print(f"結合されたファイルは '{output_file}' に保存されました。")
```
